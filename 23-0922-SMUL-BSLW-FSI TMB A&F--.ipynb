```python
# Import spaCy library
import spacy

# Import the visualisation module for rendering named entities
from spacy import displacy

# Import the Span class to represent sequences of tokens (in a document)
from spacy.tokens import Span

# Load the spaCy model and assign it to a variable 'nlp'
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "Apple Inc. invested $1 billion in Chinese ride-hailing service Didi Chuxing."

# Process the text with spaCy
doc = nlp(text)

# Adjust the label of any entity in the document that matches "Didi Chuxing" to "ORG"
# Update the named entities in the document, setting "Didi Chuxing" as an "ORG"
doc.ents = [Span(doc, ent.start, ent.end, label="ORG") if ent.text == "Didi Chuxing" else ent for ent in doc.ents]
# doc.ents = [Span(doc, aaa.start, aaa.end, label="ORG") if aaa.text == "Didi Chuxing" else aaa for aaa in doc.ents]

# Visualise the named entities
displacy.render(doc, style="ent", jupyter=True)
```

```python
# 'doc.ents' refers to the named entities identified in the doc object.
# 'for ent in doc.ents' means for each entity (ent) in the original list of entities (doc.ents).
# 'Span(doc, ent.start, ent.end, label="ORG")' is a call to create a new Span object.
# 'doc' is the document to which the span belongs.
# 'ent.start' is the start index of the entity in the document.
# 'ent.end' is the end index of the entity in the document.
# 'label="ORG"' specifies the label of the span. In this case, the label is set to "ORG", which stands for "Organization".
# 'if ent.text == "Didi Chuxing"' checks if the text of the entity (ent.text) is equal to "Didi Chuxing".

# The result is a new list of entities where every occurrence of "Didi Chuxing" is explicitly labeled as an organization, while all other entities remain unchanged.
# This list then replaces the original entities in the doc object.
```

```python
# Title: Text Mining Basic Analysis

"""
This program preprocesses a given text by tokenizing, lemmatizing,
and removing stopwords. It then analyzes the text to display
the most common words.
"""

# Import necessary libraries
import nltk
import string
import spacy

# Download necessary NLTK resources and load spaCy model
nltk.download('stopwords', quiet=True)
nlp = spacy.load("en_core_web_sm")

# Fetch stopwords once for efficiency
STOPWORDS = set(nltk.corpus.stopwords.words('english'))

def preprocess_text(text):
    """Preprocess the text."""
    # Split text into paragraphs
    paragraphs = text.split('\n\n')

    preprocessed_tokens = []

    for paragraph in paragraphs:
        # Remove punctuation
        paragraph = ''.join([char for char in paragraph if char not in string.punctuation])

        # Tokenize and lemmatize using spaCy
        doc = nlp(paragraph)
        tokens = [token.lemma_ for token in doc if token.text.lower() not in STOPWORDS]

        preprocessed_tokens.extend(tokens)

    return preprocessed_tokens

def analyze_text(tokens):
    """Analyze and display the most common words in the text."""
    freq_dist = nltk.FreqDist(tokens)

    print("Most common words:")
    for word, frequency in freq_dist.most_common(50):
        print(f"{word}: {frequency}")

def main():
    """Main function to run the program."""
    file_path = input("Enter the path to the text file to analyze (or 'q' to quit): ")

    if file_path.lower() == 'q':
        print("Exiting the program.")
        return

    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()

        # Preprocess and analyze the text
        preprocessed_text = preprocess_text(text)
        analyze_text(preprocessed_text)
    except FileNotFoundError:
        print(f"File not found at {file_path}. Please check the file path and try again.")

if __name__ == "__main__":
    main()
```

```python
# Title: Text Mining Basic Analysis with Visualization

"""
This program preprocesses a given text by tokenizing, lemmatizing,
and removing stopwords. It then visualizes the most common words
using a bar chart.
"""

# Import necessary libraries
import nltk
import string
import spacy
import matplotlib.pyplot as plt

# Download necessary NLTK resources and load spaCy model
nltk.download('stopwords', quiet=True)
nlp = spacy.load("en_core_web_sm")

# Fetch stopwords once for efficiency
STOPWORDS = set(nltk.corpus.stopwords.words('english'))

def preprocess_text(text):
    """Preprocess the text."""
    # Split text into paragraphs
    paragraphs = text.split('\n\n')

    preprocessed_tokens = []

    for paragraph in paragraphs:
        # Remove punctuation
        paragraph = ''.join([char for char in paragraph if char not in string.punctuation])

        # Tokenize and lemmatize using spaCy
        doc = nlp(paragraph)
        tokens = [token.lemma_ for token in doc if token.text.lower() not in STOPWORDS]

        preprocessed_tokens.extend(tokens)

    return preprocessed_tokens

def visualize_common_words(tokens):
    """Visualize the most common words using a bar chart."""
    freq_dist = nltk.FreqDist(tokens)
    common_words, frequencies = zip(*freq_dist.most_common(30))

    plt.figure(figsize=(10, 6))
    bars = plt.barh(common_words[::-1], frequencies[::-1], color='skyblue')
    plt.xlabel('Frequency')
    plt.ylabel('Common Words')
    plt.title('Frequency Distribution of Common Words')

    for bar, freq in zip(bars, frequencies[::-1]):
        plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, str(freq), ha='left', va='center')

    plt.show()

def main():
    """Main function to run the program."""
    file_path = input("Enter the path to the text file to analyze (or 'q' to quit): ")

    if file_path.lower() == 'q':
        print("Exiting the program.")
        return

    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()

        # Preprocess and analyze the text
        preprocessed_text = preprocess_text(text)
        visualize_common_words(preprocessed_text)
    except FileNotFoundError:
        print(f"File not found at {file_path}. Please check the file path and try again.")

if __name__ == "__main__":
    main()
```

```python
"""
FinBERT Sentiment Analysis

This script provides a simple way to use the FinBERT model from Hugging Face's
transformers library to perform sentiment analysis on financial texts.

Usage:
    Call the `predict_sentiment` function with your desired text.

Example:
    sentiment = predict_sentiment("Apple Inc. reported quarterly earnings that were above expectations.")
    print(sentiment)
"""
#!pip install transformers
#!pip install torch

import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer

# Model and tokenizer initialization
# Globals
MODEL_NAME = "ProsusAI/finbert"
MODEL = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME)
TOKENIZER = AutoTokenizer.from_pretrained(MODEL_NAME)


def predict_sentiment(text):
    """
    Predicts the sentiment of a given text using the FinBERT model.

    Parameters:
        text (str): The text for which to predict sentiment.

    Returns:
        torch.Tensor: Probabilities for each sentiment class.
    """
    try:
        # Preprocess the text
        inputs = TOKENIZER(text, return_tensors="pt")

        # Make predictions
        outputs = MODEL(**inputs)

        # Convert logits to probabilities
        probs = torch.nn.functional.softmax(outputs.logits, dim=-1)

        return probs

    except Exception as e:
        print(f"An error occurred: {e}")
        return None


# Example Usage
if __name__ == "__main__":
    sentiment = predict_sentiment("Apple Inc. reported quarterly earnings that were above expectations.")
    print(sentiment)
```

```python
# 1. Loads the FinBERT model and tokenizer from Hugging Face's transformers library.
# 2. Reads a transcript from a specified file path, removing newline characters.
# 3. Tokenizes the transcript and divides it into chunks, each containing up to 510 tokens.
# 4. Converts each chunk of tokens to their respective IDs and feeds them to the model for prediction.
# 5. Collects the predicted probabilities for each chunk.
# 6. Averages the probabilities across all chunks.
# 7. Prints the averaged probabilities.

# !pip install transformers
# !pip install torch

from transformers import AutoModelForSequenceClassification, AutoTokenizer
import torch

# Specify the model name
model_name = "ProsusAI/finbert"

# Load the model and the tokenizer
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Read from a .txt file
with open('/content/drive/MyDrive/Colab Notebooks/23-0922--text mining/2023-May-04-Apple Earnings Call Transcript.txt', 'r') as file:
    text = file.read().replace('\n', ' ')

# Tokenize the text into chunks of 510 tokens
tokenized_text = tokenizer.tokenize(text)
chunks = [tokenized_text[i:i + 510] for i in range(0, len(tokenized_text), 510)]

# Collect probabilities for each chunk
all_probs = []

# Process each chunk
for chunk in chunks:
    # Convert chunk to token ids and make prediction
    inputs = tokenizer.convert_tokens_to_ids(chunk)
    inputs = torch.tensor([inputs])

    outputs = model(inputs)
    probs = torch.nn.functional.softmax(outputs.logits, dim=-1)

    all_probs.append(probs)

# Average the probabilities (or other aggregation methods)
average_probs = torch.stack(all_probs).mean(dim=0)

# Print the average probabilities
print(average_probs)
```

```python
# Note on "chunks = [tokenized_text[i:i + 510] for i in range(0, len(tokenized_text), 510)]"
'''

The code creates a list of chunks (or segments) from the tokenized_text list. Each chunk contains up to 510 tokens from the tokenized_text.

range(0, len(tokenized_text), 510):
This produces a sequence of numbers starting from 0, incrementing by 510, and stopping before len(tokenized_text).
Essentially, it gives the starting index for each chunk.

tokenized_text[i:i + 510]:
This extracts a slice from the tokenized_text list, starting from the index i and ending at i + 510 (exclusive).
Each slice will contain up to 510 tokens.

The first chunk will contain tokens 0 to 509.
The second chunk will contain tokens 510 to 1019.
The third chunk will contain tokens 1020 to 1199.
...

'''
```
